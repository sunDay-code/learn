```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    img = cv2.imread('assets/ppl1.jpg').astype(np.float32)/255
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
# Load_img()
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
i = load_img()
display_img(i)
```

```python
gamma = 8 # increase or decrease the brightness of the img
```

```python
result = np.power(i,gamma) # take every pixel of the img, raise the power of gamma
display_img(result)
```

```python
img = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,text='ppl1',org=(50,50),fontFace=font,fontScale=3,color=(255,0,0),thickness=1)
```

```python
display_img(img)
```

```python
kernel = np.ones(shape=(5,5),dtype=np.float32)/25
kernel
```

```python
dst = cv2.filter2D(img,-1,kernel)
display_img(dst)
```

```python
img = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,text='ppl1',org=(50,50),fontFace=font,fontScale=3,color=(255,0,0),thickness=1)
print('reset')
```

```python
blurred_default = cv2.blur(img,ksize=(5,5))
display_img(blurred_default)
```

```python
blurred_gau = cv2.GaussianBlur(img,(5,5),10)
display_img(blurred_gau)
```

```python
blurred_med = cv2.medianBlur(img,5)
display_img(blurred_med)
```
