```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("assets/scene1.jpg",0) # 0 is to read it in gray scale
```

```python
plt.imshow(img,cmap='gray')
```

```python
img.max()
```

```python
# value below the threshold value will become 0
# value above will become the maxval
ret,thresh1 = cv2.threshold(img,255/2,255,cv2.THRESH_TRUNC)
```

```python
ret
```

```python
plt.imshow(thresh1,cmap='gray')
```

```python
img = cv2.imread("assets/scene2.jpg",0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
show_pic(img)
```

```python
ret, th1 = cv2.threshold(img,180,255,cv2.THRESH_BINARY)
show_pic(th1)
```

```python
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,0)
```

```python
plt.imshow(th2)
```

```python
blended = cv2.addWeighted(src1=th1,alpha=0.6,src2=th2,beta=0.4,gamma=0)
```

```python
show_pic(blended)
```
